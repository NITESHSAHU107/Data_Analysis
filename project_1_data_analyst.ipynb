Que- what are location with the fastest roller coasters (minimum of 10) ?

```python
df.query('Location!="others"')\
.groupby('Location')['Speed_mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean')['mean']\
.plot(kind='barh',figsize=(12,5),title='Average coaster speeds by location')\

```
